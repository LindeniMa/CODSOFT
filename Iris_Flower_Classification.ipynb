```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_iris
```

# ***`Load Data`***

```python
Iris = load_iris()
```

```python
Iris.data.shape
```

# ***`Convert data to Pandas dataframe`***

```python
df = pd.DataFrame(Iris.data, columns = Iris.feature_names)
```

```python
df.head()
```

```python
df.tail()
```

```python
df['target'] = Iris.target
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.describe()
```

```python
print(df.groupby('target').size())
```

# ***`Data Visualization`***

```python
# Box and Whisker plot
df.plot(kind = 'box', figsize = (8,12), subplots = True, layout = (3,2))
```

```python
# Histogram
df.hist(figsize = (12,12), color = 'r')
plt.show()
```

```python
# Correlation Heat Map
sns.heatmap(df.corr(), annot = True, cmap = 'coolwarm')
```

```python
# Pair Plot
sns.pairplot(df)
```

```python
sns.catplot(x = 'petal length (cm)', y = 'petal width (cm)', palette = 'coolwarm', hue = 'target', data = df)
```
